```python
import pandas as pd

cities = pd.read_csv('miasta.csv')
print(cities)
```

```python
print(cities.values)
```

```python
cities.loc[len(cities.index)] = [2010, 460, 555, 405]
print(cities)
```

```python
import matplotlib.pyplot as plt
size = [i for i in range(len(cities["Rok"].values))]
plt.plot(size, cities["Gdansk"].values, 'r-o')
plt.xlabel('Rok')
plt.ylabel('Liczba ludnosci [tys.]')
plt.title('Liczba ludnosci w Gdansku')
plt.xticks(size, cities["Rok"].values)
```

```python
size = [i for i in range(len(cities["Rok"].values))]
plt.plot(size, cities["Gdansk"].values, 'r-o', label="Gdansk")
plt.plot(size, cities["Poznan"].values, 'b-o', label="Poznan")
plt.plot(size, cities["Szczecin"].values, 'g-o', label="Szczecin")
plt.xlabel('Rok')
plt.ylabel('Liczba ludnosci [tys.]')
plt.title('Liczba ludnosci w miastach Polski')
plt.xticks(size, cities["Rok"].values)
plt.legend()
```

```python
# standaryzacja
cities_cp = cities.copy().drop(columns=["Rok"])
cities_std = (cities_cp - cities_cp.mean()) / cities_cp.std()
mean = cities_std.mean()
std = cities_std.std()
print("Mean:")
print(mean)
print("Std:")
print(std)
```

```python
# normalizacja
cities_norm = (cities_cp - cities_cp.min()) / \
    (cities_cp.max() - cities_cp.min())
cities_norm_min = cities_norm.min()
cities_norm_max = cities_norm.max()
print("Min:")
print(cities_norm_min)
print("Max:")
print(cities_norm_max)
```
